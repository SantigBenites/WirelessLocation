```python
from pymongo import MongoClient

all_collections = [
    "equilatero_grande_garage",
    "equilatero_grande_outdoor",
    "equilatero_medio_garage",
    "equilatero_medio_outdoor",
    "isosceles_grande_indoor",
    "isosceles_grande_outdoor",
    "isosceles_medio_outdoor",
    "obtusangulo_grande_outdoor",
    "obtusangulo_pequeno_outdoor",
    "reto_grande_garage",
    "reto_grande_indoor",
    "reto_grande_outdoor",
    "reto_medio_garage",
    "reto_medio_outdoor",
    "reto_n_quadrado_grande_indoor",
    "reto_n_quadrado_grande_outdoor",
    "reto_n_quadrado_pequeno_outdoor",
    "reto_pequeno_garage",
    "reto_pequeno_outdoor",
]

DB_NAME   = "wifi_fingerprinting_data"
MONGO_URI = "mongodb://localhost:28910/"

client = MongoClient(MONGO_URI)
db = client[DB_NAME]

counts = {}
outdoor_count = indoor_count = garage_count = 0

# Count entries per collection
for coll_name in all_collections:
    count = db[coll_name].count_documents({})
    counts[coll_name] = count
    if "outdoor" in coll_name:
        outdoor_count += count
    if "indoor" in coll_name:
        indoor_count += count
    if "garage" in coll_name:
        garage_count += count

# Combined totals
outdoor_only   = outdoor_count
indoor_only    = indoor_count
garage_only    = garage_count
outdoor_and_indoor = outdoor_count + indoor_count
outdoor_and_garage = outdoor_count + garage_count
outdoor_indoor_and_garage = outdoor_count + indoor_count + garage_count
all_data = sum(counts.values())

# Print results
print("Counts per collection:")
for coll, c in counts.items():
    print(f"  {coll}: {c}")

print("\nCombined counts:")
print(f"  Outdoor only: {outdoor_only}")
print(f"  Indoor only: {indoor_only}")
print(f"  Garage only: {garage_only}")
print(f"  Outdoor + Indoor: {outdoor_and_indoor}")
print(f"  Outdoor + Garage: {outdoor_and_garage}")
print(f"  Outdoor + Indoor + Garage: {outdoor_indoor_and_garage}")
print(f"  All data: {all_data}")
```

```python
import pandas as pd
table_path = "runs_table.csv"

# MSE GRAPHS
table_pd = pd.read_csv(table_path)

run_name_to_name = {
    "1triangle_all_spaces_meters" : "1 triangle in all spaces meters", 
    "1triangle_2space_garage_outdoor" : "1 triangle in outdoor and garage", 
    "1triangle_1space_garage_small" : "smallest triangle in garage", 
    "1triangle_1space_outdoor_small" : "smallest triangle in outdoor", 
    "all_triangles_all_spaces" : "all triangles in all spaces", 
    "all_triangles_1space_outdoor" : "all traingles in outdoor", 
    "1triangle_all_spaces" : "1 triangles in all spaces", 
    "1triangle_1space_garage" : "1 triangles in garage", 
    "1triangle_1space_outdoor" : "1 triangles in outdoor", 
    "1triangle_1space_indoor" : "1 triangles in indoor"
    }

table_pd = table_pd.sort_values('mae_m')
table_pd["run_name"] =  table_pd["run_name"].map(run_name_to_name) 
styled = table_pd.style.background_gradient(cmap="RdYlGn_r")  # _r reverses so green=low, red=high


latex = styled.to_latex(
    caption="MLP accuracy, on different spaces and triangle shapes/sizes",
    label="tab:trinagles_spaces_table",
    position="H",
    environment="table",
    convert_css=True,
    multicol_align="c",
)

display(styled)
print(latex.replace("_", "\\_"))\
```
